Выбор набора данных и подмножества признаков

```python
from sklearn.datasets import load_iris
import pandas as pd

# Загрузка набора данных
iris = load_iris()
X = iris.data
y = iris.target

# Создание подмножества признаков (D1)
# Выбираем первые три признака
D1 = X[:, :3]
```

Снижение размерности с использованием PCA

```python
from sklearn.decomposition import PCA

# Снижение размерности до двух признаков с использованием PCA
pca = PCA(n_components=2)
D2 = pca.fit_transform(D1)
```

Снижение размерности с использованием t-SNE

```python
from sklearn.manifold import TSNE

# Снижение размерности до двух признаков с использованием t-SNE
tsne = TSNE(n_components=2, random_state=42)
D3 = tsne.fit_transform(D1)
```

Визуализация

Для визуализации, мы можем использовать matplotlib. Визуализируем D2 и D3.

```python
import matplotlib.pyplot as plt

# Визуализация D2
plt.figure(figsize=(8, 6))
plt.scatter(D2[:, 0], D2[:, 1], c=y, cmap='viridis')
plt.title('PCA')
plt.show()

# Визуализация D3
plt.figure(figsize=(8, 6))
plt.scatter(D3[:, 0], D3[:, 1], c=y, cmap='viridis')
plt.title('t-SNE')
plt.show()
```

Кластеризация и оценка качества

Для кластеризации, мы можем использовать KMeans, DBSCAN и AgglomerativeClustering из sklearn.cluster. Для оценки качества кластеризации, будем использовать метрики: Silhouette Score и Adjusted rand score.

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, adjusted_rand_score

# Кластеризация с использованием KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(D2)
labels_kmeans = kmeans.labels_
score_kmeans = silhouette_score(D2, labels_kmeans)
adjusted_rand_score_kmeans = adjusted_rand_score(y, labels_kmeans)

# Кластеризация с использованием DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(D2)
labels_dbscan = dbscan.labels_
score_dbscan = silhouette_score(D2, labels_dbscan)
adjusted_rand_score_dbscan = adjusted_rand_score(y, labels_dbscan)

# Кластеризация с использованием AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clustering.fit(D2)
labels_agg = agg_clustering.labels_
score_agg = silhouette_score(D2, labels_agg)
adjusted_rand_score_agg = adjusted_rand_score(y, labels_agg)

# Вывод результатов
print("KMeans Score:", score_kmeans, "Adjusted Rand Score:", adjusted_rand_score_kmeans)
print("DBSCAN Score:", score_dbscan, "Adjusted Rand Score:", adjusted_rand_score_dbscan)
print("Agglomerative Clustering Score:", score_agg, "Adjusted Rand Score:", adjusted_rand_score_agg)
```
